#### plt.figure() 创建画布
- 想要绘图，必须先要创建一个 figure（画布），还要有 axes（坐标系）

#### 隐式创建 figure
- 当第一次执行 plt.xxx() 画图代码时，会自动判断是否已经存在画布
- 如果没有，则自动创建一个画布，并且在该画布上自动创建一个坐标系
- 如果不设置画布，一个画布上只有一个坐标系

```python
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 2
y2 = x ** 2
y3 = np.sin(x)
plt.plot(x, y1, color='gold')
plt.plot(x, y2, color='red')
plt.plot(x, y3, color='green')
plt.show()
```

#### 显式创建 figure
- 利用 plt.figure() 手动创建画布，可以创建多个画布
- 在 plt.figure() 下面的 plt.xxx() 画图代码都会画在对应的画布上面

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.figure()
plt.plot(x, y)
plt.figure()
plt.plot(x, s)
plt.figure()
plt.plot(x, c)
plt.show()
```

#### plt.figure(num=None, figsize=None, dpi=None, facecolor=None)
- num： 画布编号或名称（数字为编号，字符串为名称），不指定则按照数字顺序
- figsize： 指定 figure 的宽和高（英寸），默认为（6.4, 4.8）
- dpi： 指定画布显示的分辨率（用数字表示大小，默认为100）
- facecolor： 背景的颜色

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.figure(num=888, dpi=50)
plt.plot(x, y)
plt.figure(num='正弦曲线', dpi=100, facecolor='gold')
plt.plot(x, s)
plt.figure(num='余弦曲线', figsize=(3.2, 2.4))
plt.plot(x, c)
plt.show()
```

#### 设置坐标轴标签

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)

plt.xlabel('x轴')
plt.ylabel('y轴', fontsize=20)
plt.show()
```

#### 设置坐标轴刻度
- plt.xticks(ticks=None, labels=None) / plt.yticks(ticks=None, labels=None)
- 设置对应轴的刻度点的位置和标签
- ticks： 刻度点的位置组成的列表（可以指定为空列表，则去掉刻度，但轴还在）
- labels： 刻度点的位置上的标签组成的列表（labels不指定，则标签显示ticks值）


```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)

plt.xticks(ticks=range(-5, 5), labels='abcdefghij')
plt.yticks([5, 10, 15, 20], 'ABCD')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = np.sin(x)
plt.figure()  # 画布1
plt.plot(x, y)

# ticks的值作为刻度点的位置, labels的值作为刻度点的位置上的标签
plt.yticks([-1, -0.8, -0.5, -0.1, 1], ["a", "b", "c", "d", "e"])
plt.figure()  # 画布2
plt.plot(x, y)

# ticks指定为空列表, 去掉刻度, 但轴还在
plt.xticks(ticks=[])
plt.figure()  # 画布3
plt.plot(x, y)

plt.axis("off")  # 关闭整个坐标体系
plt.figure()  # 画布4
plt.plot(x, y)

new_xticks = np.linspace(-4, 4, 9)
# labels参数没有指定，默认和ticks参数取一样的值
plt.xticks(ticks=new_xticks)
plt.yticks(ticks=[-1, -0.8, -0.5, -0.1, 1])
plt.show()
```

#### 设置坐标边框颜色
1. 用 plt.gca() 获取到坐标体系（矩形坐标框）
2. 调用坐标体系的 spines 属性，通过 'top'、'bottom'、'left'、'right' 参数获得指定的边框
3. 结合 set_color() 方法来指定颜色（当颜色指定为 'None' 时，表示不显示该边框）

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)

ax = plt.gca()
print(ax)
print(ax.get_xlim())
print(ax.get_ylim())
print(ax.spines)
print(ax.spines['left'])
print(ax.spines['right'])
print(ax.spines['bottom'])
print(ax.spines['top'])

# ax.spines['left'].set_color('none')
# ax.spines['right'].set_color('none')
# ax.spines['bottom'].set_color('none')
# ax.spines['top'].set_color('none')
# plt.xticks([])
# plt.yticks([])

plt.axis('off')

plt.show()
```

#### 指定边框为坐标轴
1. 用 plt.gca() 获取到坐标体系（矩形坐标框）
2. 坐标体系指定轴来调用 set_ticks_position() 方法，通过'top'、'bottom'、'left'、'right' 参数指定边框为坐标轴

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)

ax = plt.gca()
# 把坐标体系的上边框作为x轴
ax.xaxis.set_ticks_position('top')
# 把坐标体系的右边框作为y轴
ax.yaxis.set_ticks_position('right')
plt.show()
```

#### 移动坐标边框
1. 用 plt.gca() 获取到坐标体系（矩形坐标框）
2. 调用坐标体系的 spines 属性，通过 'top'、'bottom'、'left'、'right' 参数获得指定的边框
3. 结合 set_position() 方法来指定边框位置

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# 选择坐标体系的底边框, 设置位置到数据为15的地方(即y轴的15)
ax.spines['bottom'].set_position(('data', 15))
# 选择坐标体系的左边框, 设置位置到数据为0的地方(即x轴原点)
ax.spines['left'].set_position(('data', 0))
plt.show()
```

#### plt.legend() 创建图例
- plt.legend(handles, labels, loc, fontsize, edgecolor, facecolor)
- handles： 控制柄，默认是一个画布中所有线对象组成的列表
- labels： 图例标签，默认是绘图函数中 label 参数值组成的列表
- loc： 图例创建的位置（默认是loc=“best”，代表自动找最好的位置）
- fontsize： 图例字体大小（用数字大小表示）
- edgecolor： 图例边框颜色
- facecolor： 图例背景颜色

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y, label='抛物线')
plt.plot(x, s, label='正弦曲线')
plt.plot(x, c, label='余弦曲线')
plt.legend()
plt.show()

line1 = plt.plot(x, y, label='抛物线')
line2 = plt.plot(x, s, label='正弦曲线')
line3 = plt.plot(x, c, label='余弦曲线')
print(line1)
print(line2)
print(line3)
plt.legend(labels=['抛物线', '正弦', '余弦'])
plt.show()
```

```python
# plot返回Line2D对象, 装在列表里, 所以解包
line1, = plt.plot(x, y, label='抛物线')
line2, = plt.plot(x, s, label='正弦曲线')
line3, = plt.plot(x, c, label='余弦曲线')
print(line1)
print(line2)
print(line3)

plt.legend(handles=[line1, line3])

plt.show()

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)
plt.legend(labels=['抛物线', '正弦', '余弦'], loc='center left')

plt.show()
```

#### plt.text() 文字说明
- plt.text(x, y, s, size, color, ha, va)
- x：文字开始写的 x 位置
- y：文字开始写的 y 位置
- s：需要写的内容
- size：文字大小
- color：文字颜色
- ha：设置水平对齐方式，可选参数：'left', 'right', 'center' 等
- va：设置垂直对齐方式 ，可选参数： 'center' , 'top' , 'bottom' 等

```python
x = np.linspace(-5, 5, num=50)
y = x ** 2 + 1
s = np.sin(x)
c = np.cos(x)

plt.plot(x, y)
plt.plot(x, s)
plt.plot(x, c)
plt.text(0, 15, s='抛物线', color='red', size=20, ha='right', va='top')

plt.show()
```

#### plt.scatter() 画散点图

- y：形状为(n,)的数组，绘图的 y 轴数据

#### plt.bar() 画条形图
- plt.bar(x, height, width, color, edgecolor, alpha, linewidth, bottom, align)
- x： x 坐标
- height：条形的高度
- width：条形的宽度，默认是0.8
- color：条形的颜色
- edgecolor：条形边框的颜色
- alpha：颜色的透明度（0~1）
- linewidth：条形边框的宽度
- bottom：条形底边的起始位置（即y轴的起始坐标）
- align：条形的对齐方式（默认为 "center"，表示刻度和条形中心对齐，"edge" 表示刻度和条形左边对齐）

```python
x = np.arange(1, 11)
y = np.random.randint(20, 40, size=10)
print(y)
plt.bar(x, y, width=0.5, color='yellow', alpha=0.5,
        linewidth=5, bottom=10, align='edge',
        hatch='.', edgecolor='black')

plt.show()
```

```python
x = np.arange(1, 11)
h1 = np.random.randint(20, 40, size=10)
h2 = np.random.randint(15, 50, size=10)
plt.bar(x, h1, bottom=0.5)
plt.bar(x, -h2, bottom=-0.5)
plt.yticks([40, 20, 0, -20, -40], [40, 20, 0, 20, 40])

for x0, y0 in zip(x, h1):
    plt.text(x0, y0 + 0.5, s=y0, ha='center', va='top', color='white')
for x0, y0 in zip(x, h2):
    plt.text(x0, -y0 - 0.5, s=y0, ha='center', va='bottom')
plt.show()
```

#### plt.subplot() 创建子图
- plt.subplot(nrows, ncols, index)
- nrows：行数
- ncols：列数
- index：指定第几个子图（从1开始）

```python
plt.figure()
# 在画布上创建2行1列的子图，并在第1个子图中绘画
plt.subplot(2, 1, 1)
x = np.linspace(-5, 5, num=60)
s = np.sin(x)
c = np.cos(x)
plt.plot(x, s)
plt.plot(x, c)

# 在画布上创建2行3列的子图，在第4个子图中绘画
plt.subplot(2, 3, 4)
x = np.random.uniform(-5, 5, size=30)
y = np.random.uniform(-1, 26, size=30)
plt.scatter(x, y, s=100, c='yellow', marker='D',
            alpha=0.5, linewidths=2, edgecolors='black')

# 在画布上创建2行3列的子图，在第5个子图中绘画
plt.subplot(2, 3, 5)
x = np.linspace(-5, 5, num=60)
y = x ** 2 + 1
plt.plot(x, y)

# 在画布上创建2行3列的子图，在第6个子图中绘画
plt.subplot(2, 3, 6)
x = np.arange(1, 24, 2.5)
h1 = np.random.randint(20, 40, size=10)
h2 = np.random.randint(15, 50, size=10)
plt.bar(x, h1, align='edge')
plt.bar(x - 0.4, h2)

plt.show()
```

#### plt.axes() 画图中图
- plt.axes() 可以通过指定相对画布的位置和宽高在一个画布中定制多个坐标轴

```python
plt.figure()
'''
对这四个数字的说明:
0.1, 0.1表示坐标框左下角的位置, x和y坐标在画布的10%的位置
0.6, 0.6表示坐标框的宽和高为画布的60%大小
'''
plt.axes([0.1, 0.1, 0.6, 0.6])
x = np.arange(1, 24, 2.5)
h1 = np.random.randint(20, 40, size=10)
h2 = np.random.randint(15, 50, size=10)
plt.bar(x, h1, align='edge')
plt.bar(x - 0.4, h2)
plt.title('条形图')

plt.axes([0.5, 0.55, 0.4, 0.35])
x = np.linspace(-5, 5, num=60)
s = np.sin(x)
c = np.cos(x)
plt.plot(x, s)
plt.plot(x, c)
plt.title('双曲线')

plt.show()
```
